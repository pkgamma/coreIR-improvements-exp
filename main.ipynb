```python
import time as time
import numpy as np
import pandas as pd
import pyterrier as pt
if not pt.started():
  pt.init()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Iterator for msmarco passage

def msmarco_generate():
    dataset = pt.get_dataset("trec-deep-learning-passages")
    with pt.io.autoopen('/content/drive/My Drive/Colab Notebooks/collection.tsv', 'rt') as corpusfile:
        for l in corpusfile:
            docno, passage = l.split("\t")
            yield {'docno' : docno, 'text' : passage}
```

```python
# Terrier Indexing

timeStr = time.ctime()
print('Current Timestamp : ', timeStr)

indexer = pt.IterDictIndexer("./indexes/msmarco-passage")
indexref = indexer.index(msmarco_generate(), fields=['docno', 'text'], meta_lengths=[20, 4096])
index = pt.IndexFactory.of(indexref)

timeStr = time.ctime()
print('Current Timestamp : ', timeStr)

print(index.getCollectionStatistics().toString())
```
